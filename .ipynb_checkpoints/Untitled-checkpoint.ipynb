```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Set time horizon
T = 200

# Initialize arrays for state variables
x = np.zeros(T)           # output gap
pi_hat = np.zeros(T)      # inflation gap
i_hat = np.zeros(T)       # interest rate gap
pi_star = np.zeros(T)     # trend inflation

# Shock standard deviations (example values)
sigma_pi = 0.9
sigma_i = 0.5
sigma_star = 0.6

# Shock correlations (example values)
rho_pi_i = 0.2
rho_pi_star = 0.1
rho_i_star = -0.3

# Construct covariance matrix
cov_matrix = np.array([
    [sigma_pi**2, rho_pi_i*sigma_pi*sigma_i, rho_pi_star*sigma_pi*sigma_star],
    [rho_pi_i*sigma_pi*sigma_i, sigma_i**2, rho_i_star*sigma_i*sigma_star],
    [rho_pi_star*sigma_pi*sigma_star, rho_i_star*sigma_i*sigma_star, sigma_star**2]
])

# Draw shocks
v = np.random.multivariate_normal(mean=[0, 0, 0], cov=cov_matrix, size=T)

# Parameters for macro dynamics (example values)
b_pi_x, b_pi_pi, b_pi_i = 0.4, 0.5, -0.3
b_ix, b_i_pi, b_ii = 0.6, 0.1, 0.7

# Euler equation coefficients (to be calibrated or specified)
f_x, rho_x, psi = 1.02, 0.02, 0.13

# Initial conditions
x[0], pi_hat[0], i_hat[0], pi_star[0] = 0, 0, 0, 0

# Simulate the system
for t in range(1, T):
    # Trend inflation
    pi_star[t] = pi_star[t-1] + v[t, 2]

    # Inflation gap
    pi_hat[t] = b_pi_x * x[t-1] + b_pi_pi * pi_hat[t-1] + b_pi_i * i_hat[t-1] + v[t, 0]

    # Interest rate gap
    i_hat[t] = b_ix * x[t-1] + b_i_pi * pi_hat[t-1] + b_ii * i_hat[t-1] + v[t, 1]

    # Real interest rate approximation
    r_t = i_hat[t]

    # Euler equation for output gap
    if t > 1:
        x[t] = f_x * x[t-1] + rho_x * x[t-2] - psi * r_t
    else:
        x[t] = f_x * x[t-1] - psi * r_t

# Observed inflation and interest rate
pi = pi_hat + pi_star
i = i_hat + pi_star

# Visualization
plt.figure(figsize=(14, 6))
plt.subplot(3, 1, 1)
plt.plot(x, label='Output Gap $x_t$')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(pi, label='Inflation $\pi_t$')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(i, label='Interest Rate $i_t$')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
T = 200

# Initialize state variables
x = np.zeros(T)          # output gap
s = np.zeros(T)          # log surplus consumption ratio
c = np.zeros(T)          # log consumption
sc = np.zeros(T)         # surplus consumption (log scale)

# Initialize shocks to consumption
sigma_c = 0.02
eps_c = np.random.normal(0, sigma_c, T)

# Preference and habit formation parameters
gamma = 2.0              # risk aversion
theta0 = 0.87            # persistence of surplus consumption
theta1 = -0.05           # dependence on current output gap
theta2 = 0.02            # dependence on lagged output gap

# Output gap initialization
x[0], x[1] = 0, 0

# Steady-state surplus consumption values
S_bar = sigma_c * np.sqrt(gamma / (1 - theta0))
s_bar = np.log(S_bar)

# Surplus consumption sensitivity function
def lambda_s(st, s_bar):
    s_max = s_bar + 0.5 * (1 - np.exp(2 * s_bar))
    if st >= s_max:
        return 0
    else:
        return 1 / np.exp(s_bar) * np.sqrt(1 - 2 * (st - s_bar)) - 1

# Simulation loop
for t in range(2, T):
    # Output gap process (AR(1))
    x[t] = 0.95 * x[t-1] + eps_c[t]

    # Assume log consumption follows output gap
    c[t] = x[t]

    # Surplus consumption dynamics
    s[t] = (
        (1 - theta0) * s_bar +
        theta0 * s[t-1] +
        theta1 * x[t] +
        theta2 * x[t-1] +
        lambda_s(s[t-1], s_bar) * eps_c[t]
    )

    # Total log surplus consumption
    sc[t] = s[t] + c[t]

# Plot results
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(c, label='Log Consumption $c_t$')
plt.plot(x, label='Output Gap $x_t$', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.title('Consumption and Output Gap')

plt.subplot(2, 1, 2)
plt.plot(s, label='Log Surplus Consumption Ratio $s_t$')
plt.axhline(s_bar, color='red', linestyle='--', label='Steady State $\overline{s}$')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.title('Surplus Consumption Dynamics')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set simulation length
T = 200

# Initialize state variables
x = np.zeros(T)           # output gap
s = np.zeros(T)           # log surplus consumption ratio
c = np.zeros(T)           # log consumption

# Shocks
sigma_c = 0.02
np.random.seed(0)
eps_c = np.random.normal(0, sigma_c, T)

# Parameters
gamma = 2.0
theta0 = 0.87
theta1 = -0.05
theta2 = 0.02

# Steady-state surplus consumption
S_bar = sigma_c * np.sqrt(gamma / (1 - theta0))
s_bar = np.log(S_bar)

# Sensitivity function λ(s)
def lambda_s(st, s_bar):
    s_max = s_bar + 0.5 * (1 - np.exp(2 * s_bar))
    if st >= s_max:
        return 0
    else:
        return 1 / np.exp(s_bar) * np.sqrt(1 - 2 * (st - s_bar)) - 1

# Simulate macro dynamics
for t in range(2, T):
    x[t] = 0.95 * x[t-1] + eps_c[t]
    c[t] = x[t]
    s[t] = (
        (1 - theta0) * s_bar +
        theta0 * s[t-1] +
        theta1 * x[t] +
        theta2 * x[t-1] +
        lambda_s(s[t-1], s_bar) * eps_c[t]
    )

# Compute λ(s_t) and standard deviation of x
lambda_vals = np.array([lambda_s(st, s_bar) for st in s])
sigma_x = np.std(x)

# Equity risk premium: γ(1 + λ(s_t))σ²_x
rp_equity = gamma * (1 + lambda_vals) * sigma_x**2

# Bond risk premia:
#   Hedging regime: Cov(x, -i - π) < 0 → negative premium
#   Risky regime: Cov(x, -i - π) > 0 → positive premium
rp_bond_hedging = -gamma * (1 + lambda_vals) * sigma_x**2
rp_bond_risky = gamma * (1 + lambda_vals) * sigma_x**2

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(rp_equity, label='Equity Risk Premium')
plt.plot(rp_bond_hedging, '--', label='Bond Risk Premium (Hedging Regime)')
plt.plot(rp_bond_risky, ':', label='Bond Risk Premium (Risky Regime)')
plt.axhline(0, color='black', linewidth=0.5)
plt.title('Time-Varying Risk Premia from Surplus Consumption')
plt.xlabel('Time')
plt.ylabel('Risk Premium (proportional)')
plt.legend()
plt.tight_layout()
plt.show()
```
