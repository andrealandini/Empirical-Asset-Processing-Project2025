```python
import pandas as pd

```

```python
df = pd.read_csv('crsp_monthly_named.txt', sep='\t')
dd = pd.read_csv('crsp_daily_named.txt', sep='\t')
```

| Column         | Description                                                        |
|----------------|--------------------------------------------------------------------|
| `permno`       | Permanent security identifier assigned by CRSP                    |
| `date`         | Date of the return (typically end of the month)                   |
| `ret`          | Monthly return (including dividends)                              |
| `shrout`       | Shares outstanding (in thousands)                                 |
| `prc`          | Closing price (signed to indicate positive/negative return days)  |
| `primaryexch`  | Primary exchange where the stock is traded                        |
| `siccd`        | Standard Industrial Classification (SIC) code                     |
| `mktcap`       | Market capitalization (price Ã— shares outstanding)                |
| `mktcap_lag`   | Lagged market capitalization (from previous month)                |
| `exchange`     | Exchange code or name                                             |
| `industry`     | Industry classification                                           |
| `ret_excess`   | Excess return over risk-free rate                                 |
| `gvkey`        | Compustat company identifier                                      |
| `securitynm`   | Security name                                                     |
| `securitytype` | Type of security (e.g., common stock)                             |
| `ticker`       | Ticker symbol                                                     |
| `tradingsymbol`| Trading symbol, often exchange-specific                           |


```python

#df_subset = df[['date', 'permno', 'ret', 'ret_excess']]

df_subset = dd[['date', 'permno', 'ret', 'ret_excess']]

# Aggregate using pivot_table to handle duplicates
ret_wide = df_subset.pivot_table(index='date', columns='permno', values='ret', aggfunc='mean')
ret_excess_wide = df_subset.pivot_table(index='date', columns='permno', values='ret_excess', aggfunc='mean')

# Rename columns
ret_wide.columns = [f'ret_{int(col)}' for col in ret_wide.columns]
ret_excess_wide.columns = [f'ret_excess_{int(col)}' for col in ret_excess_wide.columns]

# Combine into single DataFrame
final_df = pd.concat([ret_wide, ret_excess_wide], axis=1)

# Optional: sort by date
final_df = final_df.sort_index()
```
